
## **Data Cleaning**



```python
df = pd.read_sql_query("""
Select r.review_id, u.user_id,r.hotel_id , h.hotel_name as hotel_name, r.score_overall, h.country, r.score_cleanliness, r.score_comfort, r.score_facilities, r.score_location, r.score_staff, r.score_value_for_money, u.age_group,u.user_gender, u.traveller_type, r.review_date
from reviews r
join hotels h on r.hotel_id = h.hotel_id
join users u on r.user_id = u.user_id
""",conn)
```

```python
df['review_id'] = df['review_id'].astype('int64')
df['hotel_id'] = df['hotel_id'].astype('int64')
```

```python
df['review_date'] = pd.to_datetime(df['review_date'])
```

```python
df.dtypes
```

```python
df.info()
```

```python
df=df.drop_duplicates()
```

```python
df.shape
```

```python
df.set_index('review_id', inplace = True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Top 5 Most reviewed Hotels**

```python
top_hotels = df['hotel_name'].value_counts().head(10)
```

```python
plt.figure(figsize = (12,6))
top_hotels.plot(kind = 'line', color = 'pink')
plt.title('Top 5 Most Reviewed Hotels')
plt.xlabel('Hotel Name')
plt.ylabel('Number of Reviews')
plt.show()
```

**Avg. Rating per Hotel**

```python
avg_rating = df.groupby('hotel_name')['score_overall'].mean().sort_values(ascending = False).head(10)
```

```python
plt.title('Average Rating per Hotel')
plt.xlabel('Hotel Name')
plt.ylabel('Average Rating')
plt.xticks(rotation = 45)
plt.plot(avg_rating.index, avg_rating.values)
plt.show()
```

**Most Active Users**

```python
active_users = df['user_id'].value_counts().head(10)
```

```python
plt.figure(figsize = (10,6))
active_users.plot(kind = 'bar', color = 'green')
plt.title('Most Active Users')
plt.xlabel('User ID')
plt.ylabel('Number of Reviews')
plt.xticks(rotation = 45)

# Add number of reviews above each bar
#for i, count in enumerate(active_users.values):
 #   plt.text(i, count, str(count), ha='center', va='bottom')

plt.show()
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(numeric_only = True),annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
